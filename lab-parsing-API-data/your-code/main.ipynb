## Reading the data

In the next cell, use Python to obtain data from the API endpoint you chose in the previous step. Workflow:

1. Define the initial API endpoint URL.

1. Make request to obtain data of the 1st page. Flatten the data and store it in an empty object variable.

1. Find out the total page count in the 1st page data.

1. Use a FOR loop to make requests for the additional pages from 2 to page count. Append the data of each additional page to the flatterned data object.

```python
import urllib,json
import requests
import pandas as pd
from pandas.io.json import json_normalize
import re
```

```python
headers = {'User-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36'}
    
request = requests.get(url, headers = headers)
result = request.json()
data = json_normalize(result)
data.columns
flattened_data = json_normalize(data.articles[0])
flattened_data
```

```python
total_pages = result['pagination']['page_count']

lst = []
for i in range(0,total_pages):
    #Asignamos a una variable un iterador que vaya aumentando *84 hasta llegar al total de páginas de la url.
    #Luego la pasamos en la url para que vaya sacando la información de todas las páginas
    x = i*84
    url = ('https://www.zalando.fr/api/catalog/articles?categories=promo-homme&limit=84&offset='+str(x)+'&sort=popularity')
    request = requests.get(url, headers = headers)
    result = request.json()
    data = json_normalize(result)
    flattened_data = json_normalize(data['articles'][0])
    lst.append(flattened_data)
    
data = pd.concat(lst, sort= False, ignore_index = True)

data
```

```python
#Our data is still text. Convert prices into numbers.
desc = {}
#Creamos un diccionario con las marcas como llave, al que luego le añadiremos los valores.

for i in data['name'].unique():
    desc[i]=0
```

```python
# your code
for i in data.index:
    #Sacamos los números del precio y cambiamos la coma por un punto.
    original_price = re.findall('\d\s\d+,\d+|\d+,\d+', data['price.original'][i])
    promotional_price = re.findall('\d\s\d+,\d+|\d+,\d+', data['price.promotional'][i])
    
    original_price = re.sub(",",".", *original_price)
    original_price = re.sub(" ","", original_price)
    promotional_price = re.sub(",",".", *promotional_price)
    promotional_price = re.sub(" ","", promotional_price)

    #Anadimos al diccionario la diferencia acumulada de los descuentos
    desc[data['name'][i]] = desc[data['name'][i]] + (float(original_price) - float(promotional_price))

desc
```

```python
product_max_desc = max(desc.values())
product_max_desc
```
